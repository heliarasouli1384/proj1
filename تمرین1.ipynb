```python
import matplotlib.pyplot as plt
```

```python
cities= ['Tehran','Esfehan','Mashhad','Tabriz','Shiraz','Karaj','Qom','Kermanshah','Ahvaz']
x= [51.23 ,51.40 ,59.32 ,46.17 ,52.32 ,50.56 ,50.52 ,47.05 ,48.40 ]
y=[35.41 ,32.39 ,36.19 ,38.04 ,29.36 ,35.49 ,34.38 ,34.20 ,31.18 ]
population=[16800 ,2220 ,3400 ,1700 ,1560 ,1290 ,1200 ,946 ,1300]
colors = (['red','green','blue','yellow','pink','black','orange','purple','brown'])
```

```python
for i in range(len(cities)):
    plt.scatter(x[i],y[i],s=population[i],color= colors[i],label=cities[i])
plt.xlabel("lat")
plt.ylabel("long")
plt.ylim([27,40])
plt.xlim([45,62])
plt.title("10 big cities in Iran")

plt.legend(markerscale=0.1,loc=4)
plt.show()
```
